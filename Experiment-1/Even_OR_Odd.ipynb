First, let's define a function to check if a number is even or odd:

```python
def is_even_or_odd(number):
    if number % 2 == 0:
        return "Even"
    else:
        return "Odd"

# Get input from the user
try:
    user_input = int(input("Enter a number to check if it's even or odd: "))
    print(f"The number {user_input} is {is_even_or_odd(user_input)}.")
except ValueError:
    print("Invalid input. Please enter an integer.")
```
