```python
import numpy as np

print("Please enter the elements for your 3x3 matrix, row by row.")

matrix_elements = []
for i in range(3):
    while True:
        try:
            row_str = input(f"Enter 3 space-separated numbers for row {i+1}: ")
            row = list(map(float, row_str.split()))
            if len(row) == 3:
                matrix_elements.append(row)
                break
            else:
                print("Please enter exactly 3 numbers for the row.")
        except ValueError:
            print("Invalid input. Please enter numbers only.")

user_matrix = np.array(matrix_elements)

print("\nYour 3x3 matrix is:")
display(user_matrix)

determinant_value = np.linalg.det(user_matrix)
print(f"\nThe determinant of the matrix is: {determinant_value}")
```
